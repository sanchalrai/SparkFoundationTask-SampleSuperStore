```python
data['Postal Code']=data['Postal Code'].astype('object')
```

```python
corr=data.corr()
sns.heatmap(corr,annot=True,cmap='Blues')
```

```python
data=data.drop(['Postal Code'],axis=1)
```

```python
data['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=(12,10))
data['Sub-Category'].value_counts().plot.pie(autopct='dark')
plt.show()
```

```python
data['State'].value_counts()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=data,order=data['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
data.hist(figsize=(10,10),bins=30)
plt.show()
```

```python
plt.figure(figsize=(10,8))
data['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(16,8))
ax.scatter(data['Sales'],data['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```

```python
data.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise sales in every region')
sns.barplot(x='Region',y='Sales',data=data,hue='Segment',order=data['Region'].value_counts().index)
plt.xlabel('Region',fontsize=12)
plt.show()
```

```python
data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['green','yellow'],figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
sr=data.groupby('State')[['Profit','Sales']].sum().sort_values(by='Sales',ascending=False)
sr[:].plot.bar(color=['green','yellow'],figsize=(15,8))
plt.title('Profit/Loss and Sales across states')
plt.xlabel('State')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
top_state=data['State'].value_counts().nlargest(10)
top_state
```

```python
data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['green','yellow'],figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```
